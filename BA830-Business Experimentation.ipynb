Randomization Check

```python
import pandas as pd
from statsmodels.stats.proportion import proportions_ztest
```

```python
count = controltreatment_df['Treatmeant/Control'].value_counts().sort_index()

# The number of successes (number of individuals in the treatment group)
successes = count[1]

# The total number of observations
nobs = df.shape[0]

# The value of the null hypothesis (the expected proportion of individuals in the treatment group if randomization is perfect)
# Assuming equal randomization, i.e., 50% to each group
expected_prop = 0.5

# Perform the proportions z-test
stat, pval = proportions_ztest(count=successes, nobs=nobs, value=expected_prop)

print(f'Z-statistic: {stat}')
print(f'P-value: {pval}')

# Interpretation
if pval < 0.05:
    print("We reject the null hypothesis and conclude that the randomization proportion was not as intended.")
else:
    print("We fail to reject the null hypothesis and conclude that there's no significant evidence that the randomization proportion was improper.")
```
